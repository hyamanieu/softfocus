```python
from bokeh.plotting import output_notebook, show
from bokeh.layouts import *
from bokeh.models.widgets import *
from bokeh.models import *
import pandas as pd
import os

data_dir = r"C:\Users\hy.amanieu\Documents\softfocus\tests"
plot_df = pd.read_csv(os.path.join(data_dir, 'sample_0.csv'),
                      parse_dates=True,
                      infer_datetime_format=True)
cols = plot_df.columns.tolist()
output_notebook()
x_sel = Select(title='X-Axis',value=cols[0],options=cols,name='x_sel')
y_sel = Select(title='Y-Axis',value=cols[1],options=cols,name='y_sel')
y_sel2 = Select(title='Y-Axis 2',value='None',options=cols+['None'],name='y_sel2')
plot_b = Button(label="Plot", button_type="success")

p = Plot(x_range=DataRange1d(), 
         y_range=DataRange1d(), 
         plot_height=600,
         plot_width=600,
         title=Title(text='youpi'),
         name='plot')

source = ColumnDataSource(plot_df)
dummy = p.add_glyph(ColumnDataSource({'x':[0],'y':[0]}),
                            Circle(x='x',y='y',
                                   size = 4,
                                   line_alpha=20,
                                   fill_alpha=20),name='dummy')
p.tags = ['iha']
dummy.glyph.js_on_change('size', CustomJS(args=dict(p=p),code=JScode_fetch))
x_sel.js_on_change('tags', CustomJS(args=dict(p=p),code=JScode_fetch))
controls = widgetbox(x_sel,y_sel,y_sel2)
plot_tab = Panel(child=row(column(controls,plot_b),p),name='plot_tab')
tabs = Tabs(tabs=[plot_tab])
layout = column([tabs])


#place of def update_plot
def update_plot():
    
    x_sel=plot_tab.select_one({'name':'x_sel'})
    
    dummy = p.select(name='dummy')[0]#handle invisible model
    p.renderers.clear()
    p.renderers.append(dummy)

    x_axis = LinearAxis(
                    axis_label = x_sel.value,
                    ticker=BasicTicker(desired_num_ticks =10),
                    name='x_axis')
    y_axis = LinearAxis(
            axis_label = y_sel.value,
            ticker=BasicTicker(desired_num_ticks =10),
            name='y_axis')

    ly = p.add_glyph(source,
                     Line(x=x_sel.value,y=y_sel.value),
                     name='ly')


    p.add_layout(x_axis,'below')
    p.add_layout(y_axis,'left')
    p.y_range.renderers = [ly]
    if y_sel2.value.strip() != 'None':
        y_axis2 = LinearAxis(
                            axis_label = y_sel2.value,
                            ticker=BasicTicker(desired_num_ticks=10),
                            name='y_axis2',
                            y_range_name='right_axis')

        p.add_layout(y_axis2,'right')
        p.extra_y_ranges = {"right_axis": DataRange1d()}
        ly2 = p.add_glyph(source,
                                       Line(x=x_sel.value,
                                           y=y_sel2.value,
                                           line_width=2,
                                          line_color='red',
                                           name = 'ly2'),
                            y_range_name='right_axis'
                                      )

        p.extra_y_ranges['right_axis'].renderers = [ly2]
        leg_items = [LegendItem(label=y_sel.value,
                                             renderers=[ly]),
                                 LegendItem(label=y_sel2.value,
                                           renderers=[ly2])
                              ]
    else:
        leg_items = [LegendItem(label=y_sel.value,
                                             renderers=[ly]),
                              ]
    p.add_layout(Legend(items=leg_items,
                                    location='top_right')
                             )
    
update_plot()        
show(layout)
```
